```python
from langchain_huggingface import HuggingFaceEndpoint
```

```python
from dotenv import load_dotenv
import os

# Load the .env file
load_dotenv()

# Access the token
sec_key = os.getenv("HF_TOKEN")

# Print or use the token
print(sec_key)
```

```python
import os
os.environ["HF_TOKEN"]=sec_key
```

```python
repo_id="meta-llama/Llama-3.2-3B-Instruct"
llm = HuggingFaceEndpoint(repo_id=repo_id,max_length=256, temperature=0.7, token=sec_key, trust_remote_code = True )
```

```python
llm.invoke("what is football")
```

```python
from langchain import PromptTemplate, LLMChain
question = "who won the last fifa world cup"
template = """Question: {question}

Answer: Let's think step by step."""

prompt = PromptTemplate(template=template, input_variables=["question"])
print(prompt)
```

```python
llm_chain = LLMChain(prompt=prompt, llm=llm)
print(llm_chain.invoke(question))
```

```python
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoTokenizer, pipeline, AutoModelForCausalLM
```

```python
model_id="nvidia/Llama-3.1-Nemotron-70B-Instruct-HF"
model=AutoModelForCausalLM.from_pretrained(model_id)
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
pipe= pipeline("text-generation", model=model, tokenizer=tokenizer , max_new_tokens=100)
hf= HuggingFacePipeline(pipeline=pipe)
```

```python
hf.invoke("what is cricket")
```
